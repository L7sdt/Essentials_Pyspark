```
#reading file with "inferSchema" set to false - OBS: inferSchema function get value format and convert it to the appropriate format 
df = spark \
  .read.format("csv") \
  .option("inferSchema", "False") \
  .option("header", "True") \
  .csv(file)
#display(df.limit(20))
df.show(20)
```

```
#get max value from column
from pyspark.sql.functions import max
df.select(max(df.delay)).take(1)
```

```
#filter dataframe row using filter() and where() NOTE: filter() and where() are basically the same
df.filter("delay < 2").show(20)
#df.where("delay < 2").show()
```

```
#order dataframe by column
#display(df.sort("date"))
df.sort("date").show(20)
```

```
#order by ascending or descending
from pyspark.sql.functions import desc, asc, expr
df.orderBy(expr("date asc")).show(20)
```

```
#get descriptive statistics
df.describe().show()
```

```
#iterate dataframe rows
index = 0
for i in df.collect():
  print(i)
  index += 1
  if index == 20:
    break;
  #print(i[0], i[1])
```

```
#add column to dataframe
df = df.withColumn('Nova Coluna', df['delay']+2)
df.show(10)
```

```
#remove column
df = df.drop('Nova Coluna')
df.show(10)
```

```
#rename column
df.withColumnRenamed('Nova Coluna', 'New Column').show()
```

```
#filter null values from specific row
df.filter("delay is NULL").show(20)
```

```
#filter missing values
df.filter(df.delay.isNull()).show(10)
```

```
#fill missing data with '0'
df.na.fill(value=0).show(10)
```

```
#fill missing data, but only on specific column
df.na.fill(value=0, subset=['delay']).show(10)
```

```
#fill data that contains blank space string
df.na.fill("").show(10)
```

```
# remove missing values from any column
df.na.drop().show(10)
```
